```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch
import numpy as np
```

# Punto 1

```python
# create the plot object
fig, ax = plt.subplots(figsize=(8, 8))
x1 = np.linspace(0, 1000)

# add 1st constraint:
plt.plot(x1, 800 - x1, lw=3, label=r'$x_1+x_2\geq 800$')
plt.fill_between(x1, 9999, 800 - x1, alpha=0.1)

# add 2nd constraint:
plt.plot(x1, 0 + .7 * x1, lw=3, label=r'$0.21x_1-0.3x_2\leq 0$')
plt.fill_between(x1, 9999, 0 + .7 * x1, alpha=0.1)

# add 3rd constraint:
plt.plot(x1, 0 + 3 * x1, lw=3, label=r'$0.03x_1-0.01x_2\geq 0$')
plt.fill_betweenx(x1, x1 / 3, 9999, alpha=0.1)

# add non-negativity constraints
plt.plot(np.zeros_like(x1), x1, lw=3, label=r'$x_2\geq 0$')
plt.plot(x1, np.zeros_like(x1), lw=3, label=r'$x_1\geq 0$')



# labels and stuff
plt.xlabel(r'$x_1$', fontsize=16)
plt.ylabel(r'$x_2$', fontsize=16)
plt.xlim(-0.1, 1000)
plt.ylim(-0.1, 1000)
plt.legend(fontsize=14)
plt.show()
```

# Punto 2

```python
# create the plot object
fig, ax = plt.subplots(figsize=(8, 8))
x1 = np.linspace(0, 1000)

# add 1st constraint:
plt.plot(x1, 1000 - 2 * x1, lw=3, label=r'$2x_1+x_2\leq 1000$')
plt.fill_between(x1, 0, 1000 - 2 * x1, alpha=0.1)

# add 2nd constraint:
plt.plot(x1, 800 - x1, lw=3, label=r'$x_1+x_2\leq 800$')
plt.fill_between(x1, 0, 800 - x1, alpha=0.1)

# add 3rd constraint:
plt.plot(np.ones_like(x1) * 350 , x1, lw=3, label=r'$x_1-0x_2\leq 350$')
plt.fill_betweenx(x1, 0, 350+ 0 * x1, alpha=0.1)

# add non-negativity constraints
plt.plot(np.zeros_like(x1), x1, lw=3, label=r'$x_2\geq 0$')
plt.plot(x1, np.zeros_like(x1), lw=3, label=r'$x_1\geq 0$')



# labels and stuff
plt.xlabel(r'$x_1$', fontsize=16)
plt.ylabel(r'$x_2$', fontsize=16)
plt.xlim(-0.1, 1000)
plt.ylim(-0.1, 1000)
plt.legend(fontsize=14)
plt.show()
```
